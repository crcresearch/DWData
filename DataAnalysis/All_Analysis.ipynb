```python
def comp (early,later):
  #this function compares the entries between a marketplace listings of two different time periods, to see if the listings have changed or not
  shared_entries = []
  listbeforenodup=[]
  listafternodup=[]
  shared_number=0
  totalbefore=0
  totalafter=0
  earlyx=early['name'].str.lower()
  laterx=later['name'].str.lower()
  #this looks for only UNIQUE entries, does not include counts of duplicates
  for y in laterx:
    if y not in listafternodup:
      listafternodup.append(y)
      totalafter=totalafter+1
  for x in earlyx:
    if x not in listbeforenodup:
      listbeforenodup.append(x)
      totalbefore=totalbefore+1
  #this creates the list of the number of shared unique entries on both lists
  for x in listbeforenodup:  
    for y in listafternodup:
      if x==y:
        #print (x)
        shared_number=shared_number+1
        shared_entries.append(x)

  print ("Total Unique Entries Before= ",totalbefore, ": Total Unique Entries After=", totalafter,
        "\nTotal Shared Entries=",shared_number,
         "\nShared Entries",shared_entries)
  
```

```python
def simnames (datafile1,datafile2,datafile3=pd.DataFrame(),datafile4=pd.DataFrame()):
  #compares up to 4 different vendor names and returns the names that are in common
  name1x=datafile1['vendor'].str.lower()
  if not datafile2.empty:
    name2x=datafile2['vendor'].str.lower()
    name2 = {x.replace('seller: ', '')for x in name2x}
  if not datafile3.empty:
    name3x=datafile3['vendor'].str.lower()
    name3 = {x.replace('seller: ', '')for x in name3x}
  if not datafile4.empty:
    name4x=datafile4['vendor'].str.lower()
    name4 = {x.replace('seller: ', '')for x in name4x}
  name1 = {x.replace('seller: ', '')for x in name1x}
  #This creates different sets of names to analyze from datafiles
  
  simnames=[]
  if datafile2.empty:
    for name in name1:
       simnames.append(name)
  if datafile3.empty and not datafile2.empty:
    for name in name1:
      if name in name2:
        simnames.append(name)
  if datafile4.empty and not datafile3.empty:
    for name in name1:
      if name in name2 and name in name3:
        simnames.append(name)
  if not datafile3.empty and not datafile4.empty:
    for name in name1:
      if name in name2 and name in name3 and name in name4:
        simnames.append(name)
  
  return (simnames)
```
